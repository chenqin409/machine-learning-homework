```python
def cau(a):
    add=0
    t=1
    for i in a:
        add+=i**t
        t+=1
    return add
nums=[]
while(1):
    t=input("请输入数字（输入shop停止）：")
    if(t=="shop"):
        break
    try:
        num=float(t)
        nums.append(num)
    except ValueError:
        print("输入错误")
print(cau(nums))
        
```
