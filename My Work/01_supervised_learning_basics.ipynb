```python
from sklearn.datasets import load_iris

iris_dataset = load_iris()
print("Keys of iris_dataset:\n", iris_dataset.keys())
```

```python
print(iris_dataset['DESCR'][:200] + "\n...")
```

```python
print("特征名称:\n",iris_dataset['feature_names'])
print("标签名称:\n",iris_dataset['target_names'])
```

```python
print("数据形状:",iris_dataset['data'].shape)
print("标签形状:",iris_dataset['target'].shape)
```

```python
print("前5朵花的数据:\n",iris_dataset['data'][:5])
```

```python
print("前5朵花的标签:\n",iris_dataset['target'][:5])
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(
    iris_dataset['data'],
    iris_dataset['target'],
    random_state=0
)

print("训练集形状:",X_train.shape)
print("测试集形状:",X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier(n_neighbors=1)
knn.fit(X_train,y_train)
```

```python
y_pred =knn.predict(X_test)
print("预测结果:\n",y_pred)
print("真实标签:\n",y_test)
print("测试集精度:{:.2f}".format(knn.score(X_test,y_test)))
```
